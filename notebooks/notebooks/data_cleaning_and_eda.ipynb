```python
import pandas as pd
import glob
import os


path = r"C:\Users\Blene\climate-challenge-week0\notebooks\notebooks\data"
print(f"Looking for files in: {os.path.abspath(path)}")

all_files = glob.glob(os.path.join(path, "*.csv"))
print(f"Files found: {all_files}")

li = []

# 2. THE CLEANING PIPELINE
if len(all_files) == 0:
    print("❌ ERROR: No CSV files found! Check your folder path.")
else:
    for filename in all_files:
        temp_df = pd.read_csv(filename)
        
        # Adding a bit of safety here for the filename split
        name = os.path.basename(filename).replace('.csv', '')
        temp_df['Country'] = name
        
        # Convert YEAR and DOY
        temp_df['Date'] = pd.to_datetime(temp_df['YEAR'].astype(str) + temp_df['DOY'].astype(str), format='%Y%j')
        
        li.append(temp_df)
        print(f"✅ Successfully loaded: {name}")


if li:
    df = pd.concat(li, axis=0, ignore_index=True)
    print("All data consolidated!")
```

```python
# Checking for missing values and data ranges
summary = df.describe()
missing_values = df.isnull().sum()

# Specific check for NASA POWER "No Data" flags (often -999 or 999)
anomalies = (df == -999).sum()

print("Missing Values:\n", missing_values)
print("NASA Power Flags Found:\n", anomalies)
```
